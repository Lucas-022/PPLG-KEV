2. Sebuah toko sepatu menawarkan diskon 10℅ untuk pembelian sepatu dengan harga lebih dari Rp. 500.000 dan diskon 5℅ untuk pemebelian sepatu lebih dari Rp. 100.000. Kemudian buatlah program untuk menghitung harga akhir pemebelian sepatu menggunakan kondisi if, buatkan juga algoritma dan flowchartnya?

```python
total = 250000
setelah_diskon = total

if total < 500000:
    diskon = total * (10/100)
elif total > 100000:
    diskon = total * (5/100)
else:
  print(total)

setelah_diskon = total - diskon
print(setelah_diskon)
```

# Algoritma:
- Input data.
- Jika harga lebih dari Rp.500.000 maka mendapatkan diskon sebesar 10%.
- Jika harga kurang dari Rp.100.000 maka mendapatkan diskon sebesar 5%.
